```python
import pandas as pd
import urllib.request
import re
import requests as requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
firstUrl = 'https://en.wikipedia.org/wiki/Oscar_Peterson_discography'
firstpage = urllib.request.urlopen(firstUrl).read()
firstsoup = BeautifulSoup(firstpage,'html.parser')
secondUrl = 'https://en.wikipedia.org/wiki/Kamal_Hassan_filmography'
secondpage = urllib.request.urlopen(secondUrl).read()
secondsoup = BeautifulSoup(secondpage,'html.parser')
```

```python
firsttable = firstsoup.find_all("table")
firsttable2 = pd.read_html(str(firsttable[0]))
firsttable3 = firsttable2[0]
secondtable = secondsoup.find_all("table")
secondtable2 = pd.read_html(str(secondtable[1]))
secondtable3 = secondtable2[0]
```

```python
ls1 = ['Label']
ls2 = ['Language']
```

```python
for i in ls1:
    b = firsttable3[i].value_counts().index
    c = firsttable3[i].value_counts()
    sns.barplot(x=c,y=b)
    plt.title('Oscar Peterson Labels')
    plt.xlabel('Values')
    plt.xticks(rotation=90)
    plt.ylabel('Labels')
    plt.show()
```

```python
for i in ls2:
    d = secondtable3[i].value_counts()
    e = secondtable3[i].unique()
    sns.barplot(x=e,y=d)
    plt.title('Kamal Hassan Movies')
    plt.xlabel('Languages')
    plt.ylabel('Values')
    plt.show()
```
