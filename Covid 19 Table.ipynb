```python
import pandas as pd
import urllib.request
import requests as requests
from bs4 import BeautifulSoup
```

```python
firstUrl = "https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers"
firstpage = urllib.request.urlopen(firstUrl).read()
soup = BeautifulSoup(firstpage,'html.parser')
```

```python
table = soup.find("table")
table2 = pd.read_html(str(table))
table3 = table2[0]
table3.head()
table3 = table3.set_index('Rank')
```

```python
table3
```

```python
ls = ['Language family','Branch']
for i in ls:
    print(i)
    print("")
    print(table3[i].value_counts().reset_index())
    print("")
```

```python
lsa = ['Indo-European','Afroasiatic','Sino-Tibetan']
for i in lsa:
    print(i)
    print("")
    table4 = table3[table3['Language family'] == i]
    print(table4['Branch'].value_counts())
    print("")
```
