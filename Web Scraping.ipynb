# Welcome to Web Scraping

```python
import pandas as pd
import urllib.request
import requests as requests
from bs4 import BeautifulSoup
```

## Obtain the first table and repeat

Here is the information: [basketball](https://basketball.realgm.com/ncaa/stats/2019/Averages/Qualified/All/Season/All/points/desc/1/)

```python
firstUrl = "https://basketball.realgm.com/ncaa/stats/2019/Averages/Qualified/All/Season/All/points/desc/1/"
firstpage = urllib.request.urlopen(firstUrl).read()
soup = BeautifulSoup(firstpage)
container = soup.find(class_="main-container")
container2 = container.find(class_="main wrapper clearfix container")
table = container2.find("table")
first = []
for row in table.find_all("tr"):
    second = []
    for cell in row.find_all(["th","td"]):
        text = cell.text
        second.append(text)
    first.append(second)
College1 = pd.DataFrame.from_dict(first)
new_header = College1.iloc[0] 
College1 = College1[1:] 
College1.columns = new_header
del College1['#']
College1.head()
```
